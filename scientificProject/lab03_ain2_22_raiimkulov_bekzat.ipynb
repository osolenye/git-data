```python
import math

# 2.	Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

x_nums = np.linspace(-10, 10, 100)

y_nums = []
def quad_func(x):
    return x*x
for i in x_nums:
    y_nums.append(quad_func(i))
plt.plot(y_nums)
plt.xlabel('x axis')
plt.ylabel('y axis')

red_patch = mpatches.Patch(color='red', label='quad func')
plt.legend(handles=[red_patch])

plt.show()
```

```python
# 3.	Построить график произвольной гармонической функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры
import math

x_nums = np.linspace(-10, 10, 100)

y_nums = []
def harmonic_func(x):
    return math.sin(x)
for i in x_nums:
    y_nums.append(harmonic_func(i))
plt.plot(y_nums)
plt.xlabel('x axis')
plt.ylabel('y axis')

red_patch = mpatches.Patch(color='red', label='quad func')
plt.legend(handles=[red_patch])

plt.show()
```

```python
#4.	Вывести графики п.2 и 3 в разные поля при помощи функции subplot()
# 5.	Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.

# quad func and harmonic func y_nums
y_nums_quad = []
y_nums_harmonic = []

plt.figure()

plt.subplot(211)
y_nums = []
for i in x_nums:
    y_nums.append(quad_func(i))
    y_nums_quad.append(quad_func(i))
plt.plot(y_nums)
plt.xlabel('x axis')
plt.ylabel('y axis')
red_patch = mpatches.Patch(color='red', label='quad func')
plt.legend(handles=[red_patch])

plt.subplot(212)
y_nums = []
for i in x_nums:
    y_nums.append(harmonic_func(i))
    y_nums_harmonic.append(harmonic_func(i))
plt.plot(y_nums)
plt.xlabel('x axis')
plt.ylabel('y axis')
red_patch = mpatches.Patch(color='red', label='harmonic')
plt.legend(handles=[red_patch])
plt.show()


plt.plot(x_nums, y_nums_quad, x_nums, y_nums_harmonic)
red_patch = mpatches.Patch(color='red', label='both')
plt.legend(handles=[red_patch])
plt.show()
```

```python
# 6.Построить график функции, заданной параметрически,
# Параметр t от 0 до 4π
t = np.linspace(0, 4 * np.pi, 500)

x = np.cos(3 * t)
y = np.sin(2 * t)

plt.plot(x, y)

# Настройки графика
plt.title('Параметрический график: x(t) = cos(3t), y(t) = sin(2t)')
plt.xlabel('x(t) = cos(3t)')
plt.ylabel('y(t) = sin(2t)')

# Отображение графика
plt.show()
```

```python
#8.	Построить круговой график по данным п. 7
plt.style.use('_mpl-gallery-nogrid')


# make data
# x = [1, 2, 3, 4]
x = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
from matplotlib import cm

plt.style.use('_mpl-gallery')

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
